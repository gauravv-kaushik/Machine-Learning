```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.random.randn(100,1)
X
```

```python
y = (X-5)**2+ 3+np.random.randn(100,1)
```

```python
plt.scatter(X,y)
```

```python
for i in range(100):
    y = (X-5)**2+ 3+np.random.randn(100,1)
    plt.scatter(X,y)
print(y)
```

```python
B =np.c_[np.ones((100,1)),X]
print(B.shape, y.shape)
```

```python
theta = np.random.randn(2,1)
theta
```

```python
m =100
n = 0.1 # learning rate
for i in range(100):
    grad = 2/m*B.T.dot(B.dot(theta)-y)
    theta = theta - n*grad
    print(theta)
```

```python
intercept,slope =theta
```

```python
intercept
```

```python
slope
```

```python
for i in range(len(X)):
    yp = slope*X[i] + intercept
    print(yp)
```

```python
yp = slope*X + intercept
plt.plot(X,yp)
plt.scatter(X,y)
plt.show()
```
