```python
import numpy as np
import pandas as pd
```

```python
from scipy import stats 
import  matplotlib.pyplot as plt
```

```python
df = pd.read_csv("headbrain.csv")
```

```python
df.head()
```

```python
X = df["Head Size(cm^3)"].values
Y  = df["Brain Weight(grams)"].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
mean_x
mean_y
```

```python
y = 0
c=0
x =0 
for i in range(len(X)):
    x += ((X[i] - mean_x)*(Y[i]-mean_y))
    y +=((X[i] - mean_x)**2)
m=x/y       
c = mean_y-m*mean_x
print(m)
print(c)
```

```python
y1 = m*X+c
y1
```

```python
plt.plot(X,y1,color = 'r')
plt.scatter(X,Y)
plt.xlabel("HEAD SIZE IN CM^3")
plt.ylabel("Brain Weight in grams ")
plt.title("Head_size: Brain")
plt.legend(["Head Size", "brain Weight "])
```

```python
slope, intercept, r, p, se = stats.linregress(X, Y)
```
